```python
messages.rename(columns={'v1': 'Label', 'v2': 'message'}, inplace = True)
```

```python
messages.Label.replace({'ham':0,'spam':1}, inplace=True)
```

```python
messages.message = messages.message.str.lower()
```

```python
from nltk.corpus import stopwords
```

```python
l1 =  stopwords.words('english') # this shows the stopwords.
```

```python
import string
string.punctuation
```

```python
# We will remove Stopwords and Punctuations 
def text_process(mess):      ### creating a Function
    """                      # a docstring
    1. remove the punctuation
    2. remove the stopwords
    3. return the list of clean textwords
    """
    nopunc = [char for char in mess if char not in string.punctuation]
    nopunc = "".join(nopunc)
    
    return [word for word in nopunc.split() if word not in l1]
```

```python
messages['message'].apply(text_process)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
messages_count = CountVectorizer(analyzer= text_process).fit(messages['message'])
messages_count.vocabulary_
# we can see which word appears how many times in corpus of document
```

```python
df_x = messages_count.transform(messages['message'])
# Created the TDM
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df_x, messages.Label, test_size= .2)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import confusion_matrix, accuracy_score
nb = MultinomialNB()
nb.fit(x_train, y_train)
pred = nb.predict(x_test)
tab_nb = confusion_matrix(pred, y_test)
tab_nb
```

```python
accuracy_score(pred, y_test)
```
